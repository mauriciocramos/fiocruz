## Agrupando os dados

#### Calcular médias agrupada por município

```python
df_novo.groupby(['municipio']).mean().head()
```

#### Calcular idade média agrupada por município

```python
df_novo.groupby(['municipio'])[['idade_mae']].mean().head()
```

#### Calcular idade média, agrupada por município, e listada em ordem descendente.

```python
df_novo.groupby(['municipio'])[['idade_mae']].mean().sort_values('idade_mae',ascending=False).head()
```

## Pivot_Table

[documentação Pivot table](https://pandas.pydata.org/pandas-docs/stable/generated/pandas.pivot_table.html)

---

```python
df_nasc.columns
```

```python
df_nasc.pivot_table(values=['idade_mae'],index = 'consultas',columns='tipo_parto')
```

#### Adicionando highlight

```python
df_nasc.pivot_table(
                    values=['idade_mae'],
                    index = 'duracao_gestacao',
                    columns='tipo_parto', 
                    aggfunc='count') \
       .style.highlight_max(axis=1)
```

### Colocando filtro

```python
filtro = df_nasc['tipo_parto']!='Ignorado'

df_nasc[filtro].pivot_table(
                    values=['idade_mae'],
                    index = 'duracao_gestacao',
                    columns='tipo_parto', 
                    aggfunc='count') \
               .style.highlight_max(axis=1)
```

## Merge
